```python
import pandas as pd
import os
import joblib
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from google.colab import drive

# 1. Montar o Google Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Carregar a base de dados (Certifique-se de que o caminho do CSV está correto)
# Exemplo: base = pd.read_csv('/content/drive/MyDrive/.../seu_arquivo.csv')
# Vou assumir que a variável 'base' já foi carregada anteriormente,
# mas vamos garantir que ela esteja limpa:

# Limpar valores nulos (Resolve o erro de NaN)
base.dropna(inplace=True)

# 3. Definição de X e y (Baseado na sua imagem, coluna 7 é o alvo)
y = base.iloc[:, 7].values
X = base.drop(base.columns[7], axis=1).values

# 4. Transformar texto em números (LabelEncoder)
labelencoder = LabelEncoder()
for i in range(X.shape[1]):
    if X[:,i].dtype == 'object' or isinstance(X[0,i], str):
        X[:,i] = labelencoder.fit_transform(X[:,i].astype(str))

# 5. Divisão entre treinamento e teste (opcional, mas recomendado)
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.3, random_state=0)

# 6. Criação e Treinamento do Modelo
modelo = GaussianNB()
modelo.fit(X_treinamento, y_treinamento)

# 7. Salvar o modelo no Drive
pasta = "/content/drive/MyDrive/Colab Notebooks/Curso completo inteligencia artificial/3.Algoritmos de Machine Learning/NaiveBayes"
os.makedirs(pasta, exist_ok=True)
caminho_final = os.path.join(pasta, "NaiveBayes.pkl")

joblib.dump(modelo, caminho_final)

print("---")
print("✅ PROCESSO CONCLUÍDO COM SUCESSO!")
print(f"Modelo treinado e salvo em: {caminho_final}")
```

```python
previsoes = modelo.predict(X_teste)
```

```python
previsoes
```

```python
accuracy = accuracy_score(y_teste,previsoes)
precision = precision_score(y_teste, previsoes, average='weighted')
recall = recall_score(y_teste, previsoes, average='weighted')
f1 = f1_score(y_teste, previsoes, average='weighted')
print(f'Acurácia: {accuracy}, Precisão: {precision}, Recall: {recall}, F1: {f1}')
```

```python
report = classification_report(y_teste, previsoes)
print(report)
```

```python
import pandas as pd
import os
import joblib
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from google.colab import drive

# 1. Montar o Drive (pode pular se já estiver montado, mas não faz mal repetir)
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# --- CONFIGURAÇÃO ---
# Se a variável 'base' não estiver na memória, descomente a linha abaixo e coloque o caminho do seu CSV:
# base = pd.read_csv('/content/seu_arquivo.csv')

# 2. Limpeza Crítica (Resolve o erro "Input contains NaN")
base.dropna(inplace=True)

# 3. Definição de X e y (Coluna 7 como alvo, conforme suas imagens)
y = base.iloc[:, 7].values
X = base.drop(base.columns[7], axis=1).values

# 4. LabelEncoder (Transforma texto em números para o Naive Bayes)
labelencoder = LabelEncoder()
for i in range(X.shape[1]):
    if X[:,i].dtype == 'object' or isinstance(X[0,i], str):
        X[:,i] = labelencoder.fit_transform(X[:,i].astype(str))

# 5. Divisão de Treino e Teste
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=0.3, random_state=12)

# 6. Treinamento do Modelo (Cria a variável 'modelo')
modelo = GaussianNB()
modelo.fit(X_treinamento, y_treinamento)

# 7. Salvamento do Modelo no seu Drive
pasta = "/content/drive/MyDrive/Colab Notebooks/Curso completo inteligencia artificial/3.Algoritmos de Machine Learning/NaiveBayes"
os.makedirs(pasta, exist_ok=True)
caminho_arquivo = os.path.join(pasta, "NaiveBayes.pkl")

joblib.dump(modelo, caminho_arquivo)

print("---")
print(f"✅ Sucesso! Modelo treinado e salvo em: {caminho_arquivo}")
```
