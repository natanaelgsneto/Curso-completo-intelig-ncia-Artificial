```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

def treinar_e_avaliar():
    # 1. Carrega o dataset
    wine = load_wine()
    X = wine.data
    y = wine.target

    # 2. Separa os dados em treino e teste
    X_treino, X_teste, y_treino, y_teste = train_test_split(
        X, y, test_size=0.3, random_state=1
    )

    # 3. Treina o modelo
    modelo = GaussianNB()
    modelo.fit(X_treino, y_treino)

    # 4. Faz as previsões
    y_pred = modelo.predict(X_teste)

    # 5. Calcula as métricas
    acc = accuracy_score(y_teste, y_pred)
    prec = precision_score(y_teste, y_pred, average='weighted')
    rec = recall_score(y_teste, y_pred, average='weighted')
    f1 = f1_score(y_teste, y_pred, average='weighted')

    # 6. Retorna o dicionário estruturado
    resultado = {
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "n_treino": len(X_treino),
        "n_teste": len(X_teste),
        "classes": [int(c) for c in np.unique(y)]
    }

    return resultado

# --- Execução no Colab ---
# Aqui nós chamamos a função e exibimos o dicionário na tela
resultado_final = treinar_e_avaliar()
resultado_final
```
